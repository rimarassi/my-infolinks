```python
populacao_estado = pd.read_excel('caminho do arquivo POP2022_Municipios', header = 1, skipfooter = 34)
populacao_estado
```

```python
populacao_estado.groupby('UF').sum(numeric_only=True)
```

```python
populacao_estado[populacao_estado['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
populacao_estado = populacao_estado.assign(populacao_sem_parenteses = populacao_estado['POPULAÇÃO'].replace('\(\d{1,2}\)', '', regex = True),
                                           populacao = lambda x : x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex = True))
```

```python
populacao_estado[populacao_estado['POPULAÇÃO'].str.contains('\(', na = False)]
```

```python
populacao_estado = populacao_estado.astype({'populacao' : 'int64'})
```

```python
populacao_estado = populacao_estado.groupby('UF')[['populacao']].sum().reset_index()
populacao_estado
```

```python
emissoes_estado = emissoes_por_ano[emissoes_por_ano['Ano'] == 2021].groupby('Estado')[['Emissao']].sum().reset_index()
emissoes_estado
```

```python
dados_agrupados = pd.merge(emissoes_estado, populacao_estado, left_on = 'Estado', right_on= 'UF')
dados_agrupados
```

```python
dados_agrupados.plot(x = 'populacao', y = 'Emissao', kind = 'scatter', figsize = (10, 6))
```

```python
import plotly.express as px
```

```python
px.scatter(dados_agrupados, x= 'populacao', y= 'Emissao', text= 'Estado', opacity = 0)
```

```python
dados_agrupados = dados_agrupados.assign(emissao_per_capita = dados_agrupados['Emissao']/dados_agrupados['populacao']).sort_values('emissao_per_capita', ascending = False)
dados_agrupados
```

```python
px.bar(data_frame = dados_agrupados, x = 'Estado', y = 'emissao_per_capita')
```

```python
px.scatter(data_frame = dados_agrupados, x = 'populacao', y = 'Emissao', text = 'Estado', size = 'emissao_per_capita')
```
